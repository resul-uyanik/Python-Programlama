```python
# Geriye değer döndüren fonksiyonlar

# bir fonksiyonun geriye değer döndürüp döndürmediği fonksiyonda kullanılan
# return anahtar kelimesine göre değerlendirilir

# Bir fonksiyonda return anahtar kelimesi kullanılıyorsa o fonksiyon geriye
# değer döndüren bir fonksiyondur.

def toplam(a,b):
    return a+b

print(toplam(9,1))

sonuc = toplam(9,1)
print(sonuc)
print(f"Sonucun iki katı = {sonuc*2}")

def my_control(param1):
    if param1%2==0:
        return True
    else:
        return False

my_result_1 = my_control(sonuc)
print(my_result_1)

def gecis_yap():
    print("Yetkilendirmeniz olumlu. Yönlendiriliyorsunuz")

def my_control_2(param2):
    if param2:
        gecis_yap()
    else:
        print("Yetkilendirme Olumsuz")

my_control_2(my_result_1)
```
