```python
# Parametreli fonksiyon örneği
# parametre olarak verilen bir sayıdan başlayarak,
# yine parametre olarak verilen diğer sayıya kadar olan tüm sayıları 
# yazdıran fonksiyon

def my_func_1(sayi_1,sayi_2):
    for i in range(sayi_1,sayi_2+1):
        print(i)

num1 = 10
num2 = 18
my_func_1(num1,num2)

print("-"*50)

my_func_1(10,18)
```

```python
# Normal kod ile 1'den 10'a kadar olan sayıların toplamını veren kod (10'da dahil)

toplam=0
for sayi in range(1,11):
    print(f"Sayı = {sayi}\t Toplam = {toplam}+{sayi} = {sayi+toplam}")
    toplam = toplam + sayi
```

```python
# Parametre olarak girilen 2 sayı arasındaki tüm sayıların toplamını
# döndüren fonksiyon

def my_func_2(num_1, num_2):
    toplam=0 # toplama işlemini yapacak değişkeni oluşturduk
    #0 , toplama işleminde etkisiz eleman olduğu için başlangıç değeri olarak atadık
    for sayi in range(num_1, num_2+1):
        toplam = toplam + sayi
    print(f"{num_1} ve {num_2} sayıları arasındaki tüm sayıların toplamı = {toplam}")

my_func_2(-10,10)
```

```python
# Parametre olarak verilen bir listedeki a ve c harfi ile başlayan
# değerleri silen kod

my_list_3 = ["aslan","calıkuşu","karga","atmaca","baykuş","civciv"]

print(f"Listenin ilk hali = {my_list_3}")
# listenin elemanlarına tek tek erişmek için for döngüsü açıyoruz:
# boş bir liste oluşturup, silinecek elemanları orada tutuyoruz
bos_liste = []

for harf in my_list_3:
    if harf[0]=="a" or harf[0]=="c":
        bos_liste.append(harf)

print(bos_liste)
```

```python
# Şimdi de silinecek elemanların tutulduğu bos_liste'nin içine girip, ilk listemizdeki elemanlar ile bu listedeki
# elemanları karşılaştırıyoruz.

# Eğer silmek istediğimiz elemanlar birbiriyle uyuşursa o zaman silme işlemini gerçekleştireceğiz

for eleman in bos_liste:
    if eleman in my_list_3:
        my_list_3.remove(eleman)

print(my_list_3)
```

```python
# Şimdi de bu kodumuzu bir fonksiyon haline getiriyoruz

my_list_3 = ["aslan","calıkuşu","karga","atmaca","baykuş","civciv"]

def eleman_sil(liste_orijinal:list):
    bos_liste = []

    for harf in liste_orijinal:
        if harf[0]=="a" or harf[0]=="c":
            bos_liste.append(harf)
    
    for eleman in bos_liste:
        if eleman in liste_orijinal:
            liste_orijinal.remove(eleman)
    print("Silme işlemi Tamamlandı")
    print(f"Yeni Liste = {liste_orijinal}")

print(f"Listenin ilk hali = {my_list_3}")
print("-"*50)
#Şimdi de fonksiyonumuza parametre olarak, my_list_3'ü verip çalıştırıyoruz
eleman_sil(my_list_3)
```
