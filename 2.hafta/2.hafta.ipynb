```python
# 4 temel veri tipi vardır:
# Nümerik olanlar için:
# 1) tamsayılar için int (varsayılan 64 bit)
# 2) ondalıklı sayılar için float (varsayılan 64 bit)
# 3) metin/karakter tabanlı => str
# 4) mantıksal tabanlı => bool
# Değişkenlere değer ataması yapılmadan önce veri tipi tanımlamak zorunlu değil

tamsayi = 16
ondalikli_sayi = 560.89
metinsel_ifade = 'Furkan'
metinsel_ifade_2 = "ATLAN"
mantiksal_ifade = True
# Python'da bir ifadenin çıktısını yansıtmak/ekranda göstermek için
# print() fonksiyonu kullanılır

# print() fonksiyonu ile birşey gösterilirken eğer formatlama yapılmak 
# isteniyorsa print(f"statik ifade {dinamik_ifade}")

print(f"int tipindeki değer ataması = {tamsayi}")
print(f"float tipindeki değer ataması = {ondalikli_sayi}")
print(f"str tipindeki değer ataması-1 = {metinsel_ifade}")
print(f"str tipindeki değer ataması-2 = {metinsel_ifade_2}")
```

```python
# 2 sayıyla 4 temel işlemi yapın. Bölme işleminin sonucu
# hem int hem de float olarak verilecek. Ayrıca, her iki
# sayının da üssünü alın.
# Örnek gösterim:
# print(f"toplama işleminin sonucu = {sayi_1+sayi_2}")
sayi_1 = int(input("Lütfen bir tamsayı giriniz: "))
sayi_2 = int(input("Lütfen bir tamsayı giriniz: "))

print(f"toplama işleminin sonucu = {sayi_1+sayi_2}")
print(f"çıkarma işleminin sonucu = {sayi_1-sayi_2}")
print(f"çarpma işleminin sonucu = {sayi_1*sayi_2}")
print(f"bölme işleminin sonucu (float) = {sayi_1/sayi_2}")
print(f"bölme işleminin sonucu (int) = {sayi_1//sayi_2}")
print(f"1. sayının karesi = {sayi_1**2}")
print(f"2. sayının karesi = {sayi_2**2}")
print(f"{sayi_1} sayısının 6 ile bölümünden kalan = {sayi_1%6}")
print(f"{sayi_2} sayısının 6 ile bölümünden kalan = {sayi_2%6}")
```

```python
#bool tipindeki veriler için örnek
mantiksal_1 = True
mantiksal_2 = False

# int tipinde 2 değeri bool tipinde 2 değere dönüştürmek
# int 1 değeri True olacak
# int 0 değeri False olacak
# dönüşüm işlemi için şu formül izlenir
# dönüştürülmek istenen tip(dönüştürülmek istenen değer)
new_true = bool(1)
new_false = bool(0)

print(f" int 1 değeri için bool değeri karşılığı = {new_true}")
print(f" int 0 değeri için bool değeri karşılığı = {new_false}")
```

```python
# Her bir veri için tür (tip) dönüşümü örnekleri yapınız:
            
print(f"int->str str(15)= {str(15)}")
print(f"int->float float(155)= {float(155)}")
print(f"int->bool str(0)= {bool(0)}")
            
print("-"*50)
            
print(f"float->str str(15.56)= {str(15.56)}")
print(f"float->int int(85.88)= {int(85.88)}")
print(f"float->bool bool(int(1.0)= {bool(int(1.0))}")
            
print("-"*50)
            
print(f"str->int int('100')= {int('100')}")
print(f"str->float float('906.83')= {float('906.83')}")
print(f"str->bool bool(int('1')= {bool(int('1'))}")
            
print("-"*50)
            
print(f"bool->int int(True)= {int(True)}")
print(f"bool->float float(int(False)= {float(int(False))}")
print(f"bool->bool str(True)= {str(True)}")
```

```python
# str fonksiyonları
# len() fonksiyonu ile str tipindeki verinin ya da veri yapılarının
# eleman sayısını öğrenebiliriz
metinsel_ifade = "Burdur Mehmet Akif Ersoy Üniversitesi ZTYO YBS"
print(f"{metinsel_ifade}\n cümlesi toplam {len(metinsel_ifade)} karakter uzunluğundadır")

print("str özel karakterler: \n ile yeni bir satıra geçeriz (new line)")
print("str özel karakterler: bir tab boşluk \t bırakmak için kullanılır (t harfi)")

print(f"metinsel_ifade.count('i') fonksiyonu ile cümlede kaç tane i harfi geçtiği bulunur = {metinsel_ifade.count('i')}")
print(f"metinsel_ifade.find('r') ile r harfinin indeks numarasını getirir (sadece str ile kullanılır) = {metinsel_ifade.find('r')}")
print(f"metinsel_ifade.index('r') ile r harfinin indeks numarasını getirir = {metinsel_ifade.index('r')}")
```

```python
# str indeksleme
print(f"str tipindeki verilerde 2 matematiksel işlem yapılabilir")
print(f"1.si toplama işlemi: 2 str ifade + simgesi ile toplanabilir")
print(f"bu işlemin çıktısı, 2 kelimenin bitişik şekilde yazılması")
isim, soyisim = "Furkan", "ATLAN"
print(f"örneğin: {isim+soyisim} gibi")
print(f"2. işlem ise çarpma. Python'da çarpma simgesi ile str bir değişkeni çarparsanız")
print(f"Sonuç, o str ifadesinin çarpma işlemi kadar yan yana yazdırılmasıdır")
print("furkan*50 işleminin sonucu furkan değerinin 50 defa yazdırılmasıdır (yan yana ve bitişik halde")
print("*"*50)
print("-"*50)
print(f"")
```

```python
# str indeksleme
# indeksleme işlemi için değişken_adı[:bitiş_indeksi]
# yukarıdaki ifade ile başlangıç_indeksinden sona kadar yazdırır
print(f"cümlenin ilk 5 harfi = {metinsel_ifade[0:5]}")
print(f"cümlenin ilk 5 harfi = {metinsel_ifade[:5]}")

# metinsel_ifade = metinsel_ifade[:] = metinsel_ifade[::]
# metinsel_ifade[başlangıç_indeksi:bitiş_indeksi]
# metinsel_ifade[başlangıç_indeksi:bitiş_indeksi: artış_miktarı]
print(f"{metinsel_ifade}")
print(f"{metinsel_ifade[:]}")
print(f"{metinsel_ifade[::]}")

# tersten sıralamak için python'da artış miktarı olarak -1 kullanılır
print(f"metinsel_ifade[::-1] = {metinsel_ifade[::-1]}")

# indeksleme işlemin sadece -1 yazarsanız son değeri getirir
print(f"metinsel_ifade[-1] = {metinsel_ifade[-1]}")
```
