```python
# Fonksiyon Tipi-1: Geriye Değer Döndürmeyen ve Parametresiz fonksiyon

# Daha önceden tanımlanmış bir görevi, daha önceden geliştirici tarafından
# verilen iç değişkenlerle halledip, genellikle dışarıya sadece bilgi verme
# amaçlı kullanılan fonksiyonlar

# Örneğin, Fonksiyon içerisinde değeri verilen değişkenlerin toplamını
# ekrana yazdıran fonksiyon

def toplama_1():
    print(f"5+10 işleminin sonucu = {5+10}")

# NOT: Geriye değer döndürmeyen fonksiyonlar, return anahtar kelimesi
# kullanılmadığı için print() ile gösterilmeye gerek duymazlar
# bir fonksiyonu çalıştırmak için adı yazılır ve yanına hemen () simgesi
# eklenir

toplama_1()
```

```python
# Fonksiyon Tipi-2: Geriye Değer Döndürmeyen ve Parametreli
# Bu fonksiyon türünde, yine ekrana bilgi yazdırmak/ kullanıcıyı bilgilendirmek
# ya da tek bir fonksiyonu çalıştırıp sonuç göstermek için kullanılır

# Geriye değer döndürmediği için fonksiyonun içindeyken return anahtar kelimesi
# kullanılmaz. Ancak, parametreli olduğu için fonksiyonu tanımlarken parantez
# içine parametreler verilir. Eğer istenirse, parametrelerin tipi de belirtilir.

# Aşağıda, kullanıcının verdiği 2 sayının toplamını ekrana yazdıran fonksiyon
# örneği yer almaktadır.

# NOT: bir fonksiyon tanımlanırken, parametre alıyorsa biz o parametrelere
# sadece o fonksiyonun içinde geçerli olmak üzere geçici birer isim veririz.
# Örneğin, aşağıda 2 sayıyı temsil etmesi için sayi_1 ve sayi_2 isminde 
# geçiçi 2 parametre tanımladık. Bunlar normal kod sayfasına yani global
# kod alanında hiçbir anlam ifade etmez. Yani, bunların geçerli olduğu tek yer
# toplama_2 fonksiyonunun kapsamıdır.

def toplama_2(sayi_1, sayi_2):
    print(f"{sayi_1}+{sayi_2} işleminin sonucu = {sayi_1+sayi_2}")

#print(sayi_2)
# Birinci kullanım, parametreleri direkt vermek
toplama_2(10,25)

# İkinci kullanım, değerleri önce bir değişkene atayıp sonra parametre olarak
# o değişkenlerin adını yazmak

number_1, number_2 = 10, 25
toplama_2(number_1,number_2)
```

```python
# 3. Kullanım, veri girişi alıp, parametreleri buna göre ayarlamak

new_num_1 = int(input("Lütfen 1. sayıyı giriniz: "))
new_num_2 = int(input("Lütfen 2. sayıyı giriniz: "))

toplama_2(new_num_1, new_num_2)
```

```python
# Fonksiyon Tipi-3: Geriye Değer Döndüren ama parametre almayan (parametresiz)

# Bu fonksiyonlar da genel olarak, önceden ayarlanmış değerler/değişkenlerle
# bir işlem yapıp, sonucu döndürmek için kullanılır.

# Geriye değer döndürdükleri için return anahtar kelimesiyle sonucu geriye
# döndürürler ki bu sonuç başka bir fonksiyonun parametresi de olabilsin ya
# da sonradan tekrar bu sonucu kullanabilelim diye.

# Örnek, önceden tanımlanmış 2 sayının toplamını geriye döndüren fonksiyon

def toplama_3():
    return 55+15


# Geriye değer döndüren fonksiyonlar, ekrana bir sonucu basmayıp, return
# ettikleri için onları görmek istiyorsanız eğer mutlaka print içerisinde 
# kullanmalısınız. Bunun da 2 farklı kullanım örneği mevcuttur.

# Kullanım-1: Direkt fonksiyonu print()  içerisinde çalıştırmak
# Bu kullanım, sadece geriye döndürülen değeri ekrana yazdırır, herhangi bir
# değişkende tutmaz. (NOT: Bu kullanım genelde pek tercih edilmez)

print("Kullanım-1")
print(toplama_3())
print("-"*50)

# Kullanım-2: Bu kullanım örneğinde de, fonksiyonun döndürdüğü sonuç, bir
# değişkene atanır ve orada saklanır. İhtiyaç duyulduğunda artık fonksiyonu
# tekrar tekrar çalıştırmak yerine, daha önce sonucunun atandığı değişkenle
# çalışılır. NOT: Genelde tercih edilen kullanım da bu yöntemdir.

sonuc_3 = toplama_3()
print("Kullanım-2")
print(sonuc_3)
```

```python
# Fonksiyon Tipi-4: Geriye Değer Döndüren ve Parametreli Fonksiyon (Gerçek
# hayatta en sık kullanılan fonksiyon tipi)

# Bu fonksiyon tipinde, bir ya da birden fazla parametre alınır. Bunların
# tipleri birbirinden farklı olabilir. Fonksiyon içerisinde belirli işlemler
# yapılır ve daha sonra bir ya da birden fazla sonuç döndürülür.

# NOT: Birden fazla sonuç döndürüleceği zaman, döndürülen sonuçların hepsinin
# aynı tipte olması gerekmez ama hepsini bir değişken (bir veri yapısı) içinde
# döndürmek lazım geldiği için  en başta hepsi tek bir yapı içerisinde
# döndürülür. Daha sonra indeksleme ya da birden fazla değişkenle atama 
# işlemi sayesinden farklı tipteki sonuçlar çekilir.

# Örnek: Parametre olarak verilen 2 not listesinde sırayla vize ve final
# notları yer almaktadır. Bu listelerde yer alan vize notunun %40 ile 
# final notunun %60'ının toplamı 50 ve üzeri olan notlar, sınavı geçmiş
# sayılacaktır. Buna göre sınavı geçen notları, bir listeye ekleyip, sonuç
# olarak da bu listeyi döndüren fonksiyonu yazınız:

# NOT: Önce, işlemden sonra sonucu ekleyeceğiniz boş bir listeyi
# fonksiyonun içinde oluşturunuz.

# Daha sonra size parametre olarak verilen VERİ YAPISININ içine for 
# döngüsüyle girip her bir elemanına erişin.

# Eğer for döngüsünde range kullandıysanız öncelikle veri yapısının
# ilgili elemanını bir değişkene atayınız:

# elemanlara eriştikten sonra yapacağınız ilk işlem, sizden istenen
# matematiksel ya da diğer işlemlerdir. Eğer bundan önce if yapısı
# kullanmanız gerekiyorsa (büyüktür, küçüktür, negatif-pozitif)
# önce if yapısı açıp onun skopuna (kapsamına) girip sonra işlem yapmalısınız.

# İşleminiz bittikten sonra (yani for döngüsüyle ve diğer tüm hesaplamalar ile)
# işiniz bittikten sonra DOĞRU KAPSAMDAYKEN (GENELLİKLE FONKSİYONUN HEMEN 
# KENDİNDEN SONRAKİ KISMIDIR yani ilk iç kısmıdır) sonucunuzu return
# etmeniz gerekir.

# SONUÇ, tek bir değişken de olabilir, bir veri yapısı da olabilir
# başka bir fonksiyondan dönen değer de olabilir. YANİ HER ŞEY SONUÇ OLABİLİR
# ve HER ŞEY RETURN EDİLEBİLİR.

# def'ten başlayıp return'e kadar giden kısım fonksiyonun TANIMLANMA kısmıdır
# fonksiyonu çalıştırmak için fonksiyonun adını yazıp hemen yanına parantez
# aç kapa simgesini eklememiz gereklidir.

def sinav_notu(liste_vize:list, liste_final:list):
    sonuc_listesi = []

    for sayi in range(len(liste_vize)):
        vize_notu = liste_vize[sayi]
        final_notu = liste_final[sayi]

        son_not = (vize_notu*0.4) + (final_notu*0.6)
        if son_not>=50:
            sonuc_listesi.append(son_not)

    return sonuc_listesi

vize_notlari = [55,87,12,23,96]
final_notlari = [28,39,58,85,77]

sinav_sonuclari = sinav_notu(vize_notlari, final_notlari)

print(f"Sınav Sonuçları aşağıdaki gibidir: ")
print(sinav_sonuclari)
```

```python
ogrenciler = {"Furkan":[85,56],"Ahmet":[52,38],"Mehmet":[28,83],
             "Tuğçe":[76,13],"Tuba":[25,37],"Aslı":[75,85],"Murat":[63,72]}

# Parametre olarak verilen bir dictionary veri yapısı içerisinde her bir
# öğrencinin aldığı not bir liste şeklinde verilmiştir. Buna göre, listenin
# ilk elemanı vize notunu, ikinci elemanı da final notunu temsil etmektedir.

# vizenin %40 ile finalin %60'ı 50 ve üzeri öğrencileri hesaplayıp
# bir listeye kaydeden ve bu listeyi döndüren fonksiyonu yazınız:
# Örneğin: Furkan 53 puan ile geçti gibi bir ifadeyle kaydedecek ya da
# Furkan 43 puan ile kaldı yazacak

def not_hesaplama(gecici_dict:dict):
    bos_liste = []
    for my_key, my_value in gecici_dict.items():
        vize_notu = my_value[0]
        final_notu = my_value[1]

        ortalama = (vize_notu*0.4)+(final_notu*0.6)

        if ortalama>=50:
            bos_liste.append(f"{my_key} {ortalama} ile GEÇTİ")
        else:
            bos_liste.append(f"{my_key} {ortalama} ile KALDI")

    return bos_liste

print(not_hesaplama(ogrenciler))
```

```python
personel = {"Ahmet":28000, "Mehmet":19500, "Enes":55000,
           "Tuğçe":72000, "Elif":41000, "Tuba":19000, 
            "Şenay":12500,"Metin":36000}

# Parametre olarak verilen dictionary veri yapısına ait bir personel bilgilerinde
# maaşı 0-20000 TL arasında olanlara %55,
# 20001-40000 TL arasında olanlara %40
# 40001-60000 TL arasında olanlara %25
# ve 60000 TL'nin üzerinde olanlara %10 zam yapıp
# yeni bir dictionary olarak döndüren fonksiyonu yazınız.

def zam_hesapla(calisanlar:dict):
    yeni_calisanlar = {}
    for isim, maas in calisanlar.items():
        if 0<=maas<=20000:
            zam_orani = 1.55

        elif 20001<=maas<=40000:
            zam_orani = 1.4

        elif 40001<=maas<=60000:
            zam_orani = 1.25

        else:
            zam_orani = 1.1

        yeni_calisanlar[isim] = maas*zam_orani
    return yeni_calisanlar

print("Orijinal Personel Maaşları")
print(personel)
print("-"*50)
print("Zam yapılmış maaşlar")
print(zam_hesapla(personel))
```
