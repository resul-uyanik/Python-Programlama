```python
# us_hesapla() fonksiyonunun sonucunu tek mi çift mi diye kontrol eden
# başka bir fonksiyon yazınız.

def control_1(num1):
    if num1%2==0:
        return "Çift Sayı"
    return "Tek Sayı"

# Birinci çalıştırma örneği
print(control_1(us_hesapla(5)))

# İkinci çalıştırma örneği
sonuc = us_hesapla(5)
print(control_1(sonuc))
```

```python
# Parametre olarak listenin kullanıldığı fonksiyon örneği-1

# Parametre olarak verilen listedeki sadece "a" ya da "z" ya da "f" harfiyle
# başlayan değerleri yeni bir listeye ekleyerek bunları döndüren fonksiyon

def ornek_list_1(param_list_1:list):
    bos_liste = []

    for aranan in param_list_1:
        if aranan[0]=="a" or aranan[0]=="f" or aranan[0]=="z":
            bos_liste.append(aranan)

    return bos_liste

my_list_1 = ["furkan","zehra","barış","kadir","alper","yusuf","tuba"]
print(ornek_list_1(my_list_1))
```

```python
# return yaparken tek bir sonuç döndürmek zorunda değiliz

# Parametre olarak verilen bir listedeki tek sayıların karesini alıp
# bir listeye, çift sayıların küpünün yarısını hesaplayıp başka bir listeye
# ekleyin
# Bu fonksiyon sonuç olarak her iki listeyi de DÖNDÜRECEK!!!

def liste_ornek_1(liste_gecici:list):
    liste_tek, liste_cift = [], []
    for sayi in liste_gecici:
        if sayi%2==0:
            liste_cift.append((sayi**3)/2)
        else:
            liste_tek.append(sayi**2)
    return liste_tek, liste_cift

# Kullanım-1: İki elemanı da birlikte ele alıp sonra bu değişkeni
# indeksleyerek her iki sonucu da ayrı ayrı yazdırmak
sonuc_liste = liste_ornek_1([15,88,25,39,77,96,55000,-236,-12,0])
print(sonuc_liste)
print("-"*50)
print(type(sonuc_liste))
print("-"*50)
print(sonuc_liste[0])
print("-"*50)
print(sonuc_liste[1])
```

```python
# Kullanım-2, en baştan iki sonuç değişkeni tanımlayıp sonuçları bunlarda tutmak

result_tek, result_cift = liste_ornek_1([15,88,25,39,77,96,55000,-236,-12,0])
print("Tek sayılar: ")
print(result_tek)

print("-"*50)

print("Çift sayılar: ")
print(result_cift)
```

```python
# parametre olarak verilen iki sayıyı birbiriyle değiştirip döndüren fonksiyon
# örneğin fonksiyona 5,6 sayıları parametre verildiğinde önce 6'yı sonra
# 5'i döndürecek

def my_swap(a,b):
    print(f"ilk sayı = {a}\t ikinci sayı = {b}")
    a,b = b,a
    print(f"ilk sayı = {a}\t ikinci sayı = {b}")
    return a,b

sayi_1, sayi_2 = my_swap(5,6)
print(sayi_1)
print(sayi_2)
```

```python
# Vize ve final notlarını iki liste içerisinde iki parametre olarak alıp
# Vize'nin %40'ını, finalin %60'ını toplayıp 50 ve üzeri olan öğrenciler
# için 'GEÇTİ', diğerleri için 'KALDI' mesajını listeye ekleyen fonksiyonu
# yazınız

ornek_vize = [45,56,38,49,12,86,38]
ornek_final = [43,39,82,55,44,70,36]

def final_hesaplama(list_vize:list,list_final:list):
    sonuc_listesi = []

    for i in range(len(list_vize)):
        vize_not, final_not = list_vize[i], list_final[i]
        ortalama = (vize_not*0.4) + (final_not*0.6)
        if ortalama>=50:
            sonuc_listesi.append('GEÇTİ')
        else:
            sonuc_listesi.append('KALDI')
    return sonuc_listesi

print(final_hesaplama(ornek_vize,ornek_final))
```

```python
# bir maaş listesini parametre olarak alıp;
# 0 ila 22000 arasındaki ücretlere %60
# 22001 ila 60000 arasındaki ücretlere %40
# 60001 ve üzeri maaşlara %15 zam yapıp, zamlı ücretleri
# bir listeye ekleyerek bu listeyi döndüren fonksiyon

ornek_maas_listesi = [18000,21520,38650,74600,102300,19500,64200,46800]
print(f"Orijinal Maaşlar = {ornek_maas_listesi}")
print("-"*50)

def maas_zam_hesapla(liste_maas:list):
    liste_zamli = []

    for ucret in liste_maas:
        if ucret>0 and ucret<=22000:
            liste_zamli.append(ucret*1.6) # %60 zamlı hali 1.6 ile gösterilir
        elif ucret>22000 and ucret<=60000:
            liste_zamli.append(ucret*1.4) # %40 zamlı hali 1.4 ile gösterilir
        elif ucret>60000:
            liste_zamli.append(ucret*1.15)
        else:
            pass
    return liste_zamli

print("Zamlı Ücretler:")
print(maas_zam_hesapla(ornek_maas_listesi))
```
