```python
# 3 farklı hata yönetimi vardır:
# 1) try-except
# 2) raise
# 3) assert ile assertionError (bu hata türü güvenlik açığı barındırır)
```

```python
# try-except ile Hata Yönetimi:
# çalışmasını istediğiniz ve hata alma olasılığı olan kodları try bloku
# içinde yazarsınız
# eğer beklediğiniz özel bir hata varsa bunu da except bloku içerisinde
# yazarsınız ve vermek istediğiniz özel bir mesaj ya da yönlendirme ifadesi
# varsa onu da yine except blokları içerisinde yazarsınız
# son olarak hata alındıktan sonra bile çalışmasını istediğiniz kodlar
# ya da onun haricinde vermek istediğiniz bir mesaj varsa (hata olsa da 
# olmasa da çalışmasını istediğiniz bir kod varsa) onu da finally blokları
# içerisinde tanımlarsınız
```

```python
try:
    sayi_1 = int(input("1. sayıyı giriniz: "))
    sayi_2 = int(input("2. sayıyı giriniz: "))

    print(sayi_1+sayi_2)

except Exception as e:
    print("Hata Oluştu")
    print(type(e).__name__)
```

```python
class HataFurkan(Exception):
    pass

print("deneme")
sayi_1 = int(input("Lütfen bir tam sayı giriniz: "))

if sayi_1<0:
    raise HataFurkan("Bizde yanlış olmaz kardeş")
```

```python
# assert ile hata yönetimi
# assert ifadesinin yanına mantıksal bir şart yazılır
# eğer bu şart sağlanıyorsa yani dönen cevap True ise kod çalışır
# değilse, assert ifadesinden sonra virgülden sonraki kısımda str olarak
# verilen ifade hata mesajı olarak kabul edilir.
# NOT: assert anahtar kelimesinin kullanıldığı kodlarda eğer hata oluşursa
# dönen hatanın tipi AssertionError'dur.

sayi_1 = int(input("lütfen bir tamsayı giriniz: "))

assert sayi_1>0, "Lütfen istenilen aralıkta bir sayı giriniz"
print(f"girmiş olduğunuz sayı = {sayi_1}")
```

```python
# assert ile hata yönetimi

sayi_1 = int(input("lütfen bir tamsayı giriniz: "))

assert sayi_1>0, "Lütfen istenilen aralıkta bir sayı giriniz"
print(f"girmiş olduğunuz sayı = {sayi_1}")
```

```python
# iki sayıyla bölme işlemi yapan bir fonksiyon yazın
# payda olarak kabul edeceğiniz sayı için assert ifadesi ile sıfırdan
# büyük olma şartını ekleyin
# eğer sorunsuz bir şekilde işlem olursa, sonucu geriye döndürün

def bolme_islemi(num_1,num_2):
    # assert'in yanındaki ifadenin cevabı True gelirse altındaki kod çalışır
    # False gelirse yanındaki hata mesajını fırlatır
    assert num_2!=0, "Lütfen ikinci sayıyı 0 girmeyiniz"
    return f"{num_1}/{num_2} = {num_1/num_2}"

print(bolme_islemi(9,3))
print(bolme_islemi(9,0))
```

```python
# Python'da try-except HATA YÖNETİMİ için kullanılır
# raise ve assert ise hata fırlatmak için kullanılır
# Bundan dolayı, ilkinde (try-except'de) kod çalışmaya devam eder
# Ancak diğerlerinde (raise ve assert) kod çalışmayı durdurmak ZORUNDADIR
# Çünkü, assert ve raise kullanımı aslında YAPILMAMASI GEREKEN ya da TOLERE
# EDİLEMEYECEK BİR İŞLEMİN YAPILDIĞINI YA DA YANITIN DÖNDÜRÜLDÜĞÜNÜ ifade eder.
```

```python
# ValueError hata sınıfına ait bir kod yazın. Şöyle ki, try-except
# kullanarak yazacağınız ve matematiksel işlem yapan bir kodda ValueError
# hatası alacaksınız. Ancak, ValueError tipinde bir hata aldıktan sonra
# "Lütfen int tipinde bir tamsayı giriniz" uyarısı vereceksiniz
# NOT: ValueError, bir veritipine uygun değer girilmediğinde ortaya çıkan
# bir hatadır. Örneğin int girilmesi gereken yerde siz hgsf gibi bir ifade
# girerseniz burada ValueError tipinde bir hata alırsınız


try:
    sayi_1 = int(input("1. sayı: "))
    sayi_2 = int(input("2. sayı: "))

    print(sayi_1-sayi_2)
except ValueError:
    print("Lütfen int tipinde bir tamsayı giriniz")
finally:
    print("Burası her halükarda çalışır")
    print(10+25)
    print("-"*50)
```

```python
# ValueError hata sınıfına ait bir kod yazın. Şöyle ki, try-except
# kullanarak yazacağınız ve matematiksel işlem yapan bir kodda ValueError
# hatası alacaksınız. Ancak, ValueError tipinde bir hata aldıktan sonra
# "Lütfen int tipinde bir tamsayı giriniz" uyarısı vereceksiniz
# NOT: ValueError, bir veritipine uygun değer girilmediğinde ortaya çıkan
# bir hatadır. Örneğin int girilmesi gereken yerde siz hgsf gibi bir ifade
# girerseniz burada ValueError tipinde bir hata alırsınız


try:
    sayi_1 = int(input("1. sayı: "))
    sayi_2 = int(input("2. sayı: "))

    print(sayi_1-sayi_2)
except ValueError:
    raise ValueError("Lütfen int tipinde bir tamsayı giriniz")
```

```python
# for döngüsüyle bir liste veri yapısı içerisine giriniz.
# listenin eleman sayısından büyük bir sayıyla indeksleme yapan bir işlem
# yapınız
# IndexError sınıfını raise ile fırlatınız
# Hata açıklamasında şu çıktıyı veriniz: "Lütfen eleman sayısı uzunluğundan"
# küçük bir değer giriniz

# bu örneğin aynısını bir de raise kullanmadan hata yönetimi çerçevesinde
# yapın

my_list_1 = [2,5,98,False,"True",75.56]

try:
    print(my_list_1[100])

except:
    raise IndexError("Lütfen eleman sayısı uzunluğundan küçük bir değer giriniz")
```

```python
# Yukarıdaki işlemi sadece try-except kullanarak hata yönetimi çerçevesinde
# ele almak

try:
    print(my_list_1[100])
except IndexError as e:
    print(f"Hata = {type(e).__name__} Mesaj = Lütfen eleman sayısı uzunluğundan küçük bir değer giriniz")
```

```python
# Aynı işlemi (indeks hatası) assert ile yapınız
# eğer indeks hatası alınacaksa, yukarıdaki mesaj ile bu hata verilecek
# Mesaj = Lütfen eleman sayısı uzunluğundan küçük bir değer giriniz

# Kontrol edeceğimiz şey listenin eleman sayısı yani uzunluğu
indeks_sayisi = 100

assert indeks_sayisi<len(my_list_1), "Lütfen eleman sayısı uzunluğundan küçük bir değer giriniz"
print(my_list_1[indeks_sayisi])
```
