```python
# Döngüler
# Python'da özellikle for döngüsü birçok işlem için kullanılır
# for döngüsünün nümerik verilerle kullanımında 3 farklı yol söz konusudur

# for döngüsü 1 değerle birlikte kullanıldığında bu değer varsayılan olarak
# bitiş değeri anlamına gelir ve sıfırdan başlar

# söz dizimi for anahtar_değer in range(bitiş_değeri):
# yukarıdaki şekliyle kullanımda sıfırdan başlar ve bitiş_değeri-1'e kadar
# tüm sayıları yazdırır

# aşağıdaki kullanımda ilk değer ve son değeri söyleyiniz
# ilk değer = 0, son değer = 99
for i in range(100):
    print(i)
```

```python
# if ile for kullanımı
# break continue anahtar kelimeleri ile örnekler
# break kullanıldığı zaman döngü kırılır/kesilir yani döngüden çıkılır
# continue kullanıldığı zaman ilgili değeri göstermez/işlem yapmaz
# ancak döngü kaldığı yerden yine devam eder

# Örnek kullanım

# tek sayıları yazdırmayan for döngüsü kodu (continue) anahtar kelimesi ile

for i in range(100):
    if i%2!=0:
        continue
    else:
        print(i)
```

```python
# eğer döngüdeki sayı 95 ise döngüyü kırsın

for i in range(100):
    if i==95:
        break
    else:
        print(i)
```

```python
# for döngüsü 2. kullanımı
# 2 değer verilir ve söz dizimi şöyledir:
# for anahtar_kelime in range(başlangıç_değeri, bitiş_değeri):
# başlangıç_değeri dahildir, bitiş_değeri dahil değildir (1 eksiği kadar çalışır)

# ilk değer = 2
# son değer = 64
for i in range(2,65):
    print(i)
```

```python
# for döngüsü 3. kullanımı

# söz dizimi aşağıdaki gibidir:

# for anahtar_kelime in range(başlangıç_değeri, bitiş_değeri, artış_miktarı):

# ilk değer = 5
# son değer = 100

for i in range(5,101,5):
    print(i)
```

```python
# 5 ile 109 arasında 
# hem 2'ye hem de 3'e aynı anda tam bölünen tam sayıları yazdıran kod

# 1. çözüm (en sağlam yol)

for i in range(5,109):
    if i%2==0 and i%3==0:
        print(i)
```

```python
# 5 ile 109 arasında 
# hem 2'ye hem de 3'e aynı anda tam bölünen tam sayıları yazdıran kod

# 2. çözüm

for i in range(5,109):
    if i%6==0:
        print(i)
```

```python
# 5 ile 109 arasında 
# hem 2'ye hem de 3'e aynı anda tam bölünen tam sayıları yazdıran kod

# 3. çözüm (artış miktarına göre)
# 6 sayısı 5'ten büyük olduğu için ve 5 ile 109 arasında yer aldığı için
# başlangıç değeri olarak seçilmesinde herhangi bir engel yoktur
# NOT: Sınavda başlangıç değeri denildiğinde bu yöntem geçerli olmaz

for i in range(6,109,6):
    print(i)
```

```python
# 3 ün ve 5'in katı olan çift sayıları yazdırın
# aralık (8-306)

# 1. yöntem (tercih edilen ve önerilen)

for i in range(8,306):
    if i%15==0 and i%2==0:
        print(i)
```

```python
# 3 ün ve 5'in katı olan çift sayıları yazdırın
# aralık (8-306)

# 2. yöntem (iç içe if)

for i in range(8,306):
    if i%15==0:
        if i%2==0:
            print(i)
```

```python
# 1 ile 901 arasında
# aynı anda 3'e ve 7'ye tam bölünen ama 6'ya tam bölünmeyen sayıları ve
# toplam kaç sayı olduğunu yazdıran kod

sayac = 0

for i in range(1,901):
    if i%21==0 and i%6!=0:
        print(i)
        sayac = sayac + 1

print(f"21 ile tam bölünen ama 6'ya tam bölünmeyen {sayac} adet sayı bulundu")
```

```python
# 1 ile 901 arasında
# aynı anda 3'e ve 7'ye tam bölünen ama 6'ya tam bölünmeyen ilk 12 sayıyı yazdırın

sayac = 0
for i in range(1,901):
    if i%21==0 and i%6!=0:
        sayac = sayac + 1
        if sayac>12:
            break
        else:
            print(f"{sayac}. değer = {i}")
```

```python
# iç içe for döngüsü kullanımı
# içte olan for döngüsünün bütün adımları tamamlandıktan sonra
# yukarıdaki (dıştaki) for döngüsünün diğer adımına geçilir

for i in range(1,3):
    for j in range(3,5):
        print(f"i={i} j={j} için i+j = {i+j}")
```

```python
# 1'den 10'a kadar olan sayılar içerisinde bir döngü tanımlayın (dış döngü)
# 11'den 20'ye kadar olan sayılar içerisinde bir döngü tanımlayın (iç döngü)

# her iki döngünün tüm elemanlarını çarpan ve sadece sonucu çift olan
# sonuçları yazdıran kodu yazınız

sayac_toplam = 0
sayac_cift = 0
for dis_deger in range(1,11):
    for ic_deger in range(11,21):
        sayac_toplam += 1
        sonuc = dis_deger * ic_deger

        if sonuc%2==0:
            sayac_cift += 1
            print(f"dış değer={dis_deger} ve ic_deger = {ic_deger} için = {sonuc}")

print(f"{sayac_toplam} adet çarpma işlemi yapıldı")
print(f"{sayac_cift} adet sonuç çift olarak bulundu")
```
