```python
# Python'da if kullanımı:

# Python'da farklı durum/olasılıkları kontrol etmek için if yapısı kullanılır
# if ile en güçlü olasılık olan durum seçilir
# elif ile diğer olasılık ya da olasılıklar kontrol edilir
# son olarak else bloku ile istediğiniz kontrolü sağlayabilirsiniz
# if ve elif içerisinde yer almayan durumu else bloku ile kontrol edebilirsiniz

# if yapısında kullanılan karşılaştırma operatörleri

# == eşittir
# != eşit değildir
# > büyüktür
# < küçüktür
# >= büyük eşittir
# <= küçük eşittir

# Kullanıcıdan iki adet tam sayı girişi alınız
# Yukarıda verilen karşılaştırma operatörleri ile bu iki sayıyı
# karşılaştırınız


# Kullanıcıdan 2 adet int tipinde giriş alıyoruz
sayi_1 = int(input("1. sayıyı giriniz: "))
sayi_2 = int(input("2. sayıyı giriniz: "))

print(f"{sayi_1}>{sayi_2} = {sayi_1>sayi_2}")
print(f"{sayi_1}<{sayi_2} = {sayi_1<sayi_2}")
print(f"{sayi_1}>={sayi_2} = {sayi_1>=sayi_2}")
print(f"{sayi_1}<={sayi_2} = {sayi_1<=sayi_2}")
print(f"{sayi_1}=={sayi_2} = {sayi_1==sayi_2}")
print(f"{sayi_1}!={sayi_2} = {sayi_1!=sayi_2}")
```

```python
# Mantıksal sorgularda 'DEĞİLİ' işlemi
# ikili sistemlerde 0 veya 1, False veya True yanıtı vardır sadece
# Bir işlemin değili, o işlemin tersi demektir aslında
# Dolayısıyla False'nin değili True, True'nin değili False
# 1'in değili 0, 0'ın değili 1

# Python'da if kullanımı:

# Python'da farklı durum/olasılıkları kontrol etmek için if yapısı kullanılır
# if ile en güçlü olasılık olan durum seçilir
# elif ile diğer olasılık ya da olasılıklar kontrol edilir
# son olarak else bloku ile istediğiniz kontrolü sağlayabilirsiniz
# if ve elif içerisinde yer almayan durumu else bloku ile kontrol edebilirsiniz

# if yapısında kullanılan karşılaştırma operatörleri

# == eşittir
# != eşit değildir
# > büyüktür
# < küçüktür
# >= büyük eşittir
# <= küçük eşittir

# Kullanıcıdan iki adet tam sayı girişi alınız
# Yukarıda verilen karşılaştırma operatörleri ile bu iki sayıyı
# karşılaştırınız


# Kullanıcıdan 2 adet int tipinde giriş alıyoruz
sayi_1 = int(input("1. sayıyı giriniz: "))
sayi_2 = int(input("2. sayıyı giriniz: "))

print(f"{sayi_1}>{sayi_2} = {not(sayi_1>sayi_2)}")
print(f"{sayi_1}<{sayi_2} = {not(sayi_1<sayi_2)}")
print(f"{sayi_1}>={sayi_2} = {not(sayi_1>=sayi_2)}")
print(f"{sayi_1}<={sayi_2} = {not(sayi_1<=sayi_2)}")
print(f"{sayi_1}=={sayi_2} = {not(sayi_1==sayi_2)}")
print(f"{sayi_1}!={sayi_2} = {not(sayi_1!=sayi_2)}")
```

```python
# Python'da and ve or işlemi istenirse simgelerle
# istenirse ingilizce and ve or kelimeleri ile yapılabilir

sayi_1 = int(input("1. sayıyı giriniz: "))
sayi_2 = int(input("2. sayıyı giriniz: "))

if sayi_1 > sayi_2 or sayi_1 == sayi_2:
    print(f"{sayi_1} sayısı {sayi_2} sayısından büyük veya eşittir")
else:
    print(f"{sayi_1} sayısı {sayi_2} sayısından küçük veya eşittir")
```

```python
# Python'da and ve or işlemi istenirse simgelerle
# istenirse ingilizce and ve or kelimeleri ile yapılabilir

sayi_1 = int(input("1. sayıyı giriniz: "))
sayi_2 = int(input("2. sayıyı giriniz: "))
sayi_3 = int(input("3. sayıyı giriniz: "))

# and işleminde True yanıtı almak istiyorsanız tüm şartlar True
# yanıtını döndürmelidir
# tek bir şart bile False yanıtı döndürürse bütün karşılaştırmaların sonucu
# False olarak döner

if sayi_3 < sayi_1 and sayi_3 < sayi_2:
    print(f"{sayi_3} sayısı en küçük sayıdır")
else:
    print(f"{sayi_3} sayısı en küçük sayı değildir")
```

```python
# Kullanıcıdan 3 adet tamsayı girişi alan
# bu sayılardan en büyüğünü bulup, yazdıran kodu yazınız

sayi_1 = int(input("1. sayıyı giriniz: "))
sayi_2 = int(input("2. sayıyı giriniz: "))
sayi_3 = int(input("3. sayıyı giriniz: "))

if sayi_1>sayi_2 and sayi_1>sayi_3:
    print(f"{sayi_1} sayısı en büyük sayıdır")

elif sayi_2> sayi_1 and sayi_2 > sayi_3:
    print(f"{sayi_2} sayısı en büyük sayıdır")

elif sayi_3> sayi_1 and sayi_3 > sayi_2:
    print(f"{sayi_3} sayısı en büyük sayıdır")

else:
    print("Tüm sayılar birbirine eşittir.")
```
