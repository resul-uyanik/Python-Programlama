```python
# List veri yapısından sonra hatta onunla birlikte en sık kullanılan built-in (yerleşik)
# veri yapısı Dictionary veri yapısıdır. Türkçedeki karşılığı sözlük olarak bilinir.

# Köşeli parantez simgesiyle tanımlanır ama ayırt edici özelliği bu değildir
# ayırt edici özelliği "key":Value şeklinde tanımlanmasıdır
# Key değişkeni eşsiz (unique) bir değişken olmak zorundadır
# value ise hepsinde aynı olsa da olur.
# Aşağıda örnek bir dictionary veri yapısı tanımlaması verilmektedir.

my_dictionary_1 = {"isim":"Furkan","soyisim":"ATLAN","dersler":["Python Programlama",
                    "Bilgi Güvenliği"],"is_yuku":[3,4]}

# bir dictionary veri yapısında sadece key değerlerini göstermek istiyorsanı
# dictionary_değişkeni.keys() fonksiyonunu çağırırsınız

# aynı şekilde sadece değerleri (values) görmek istiyorsanız 
# dictionary_değişkeni.values() fonksiyonunu çağırırsınız

# Hem key hem de ona karşılık gelen values değerlerini aynı anda görmek istiyorsanız
# yani bunlara aynı anda erişmek istiyorsanız dictionary_değişkeni.items()
# fonksiyonunu çağırırsınız

# NOT: sadece key ya da sadece values değerlerini yazdırdığınızda tek bir yapı vardır
# yani tek indeksleme yapılır

# Ancak, hem key hem de values'i aynı anda getirirken iki değişken vardır
```

```python
print(f"Sadece anahtarlar (keys) = {my_dictionary_1.keys()}")
print("-"*50)
print(f"Sadece değerler (values) = {my_dictionary_1.values()}")
print("-"*50)
print(f"Hem keys hem de values (aynı anda/birlikte) = {my_dictionary_1.items()}")
```

```python
# İçerisinde "isciler" isminde bir anahtar ve bu anahtara ait
# 5 işçinin maaşını barındıran bir liste olan dictionary tanımlayın

# Sonra, yazacağınız kod ile bu maaş listesindeki tüm maaşlara %25 zam yapan
# ve bu yeni maaşları "isciler" anahtarının yeni değeri haline getiren kodu yazın

# örneğin şöyle bir listede [100,100,100,100,100]

my_dictionary_2 = {"isciler":[15000,33266,78002,65221,43900]}

# ilk aşamada değerleri (values) bir liste olarak alıyorum:

maas_listesi = my_dictionary_2["isciler"]

# sonra her bir değeri değiştirmek için for döngüsüyle bu listenin içine giriyoruz

print(f"Orijinal Maaşlar = {my_dictionary_2['isciler']}")

for i in range(len(maas_listesi)):
    deger = maas_listesi[i] # deger = 100
    deger = deger * 1.25 # deger = 100+25 = 125
    #deger = deger + (deger*0.25)
    maas_listesi[i] = deger # 100=125 (eski değer = yeni değer ataması yaptık)

my_dictionary_2["isciler"] = maas_listesi

print(f"Zamlı Maaşlar = {my_dictionary_2['isciler']}")
```

```python
cocuklar = {"ahmet":25,"ali":18,"berfin":19,"elif":22,
           "leyla":16,"tuğçe":23,"hakan":21,"yusuf":17}

# yukarıdak verilen dictionary içerisinde yaşı 18'den küçük olanlar için
# boş bir listeye "ÇOCUK", 18 ve 18'den büyük olanlar için "YETİŞKİN" yazdıran
# kodu yazın ve listeyi yazdırın

bos_liste = []

yas_listesi = cocuklar.values()

for deger in yas_listesi:
    if deger>=18:
        bos_liste.append("YETİŞKİN")
    else:
        bos_liste.append("ÇOCUK")

print(f"Liste = {bos_liste}")
```

```python
ogrenciler = {"berfin":[76,98],"ahmet":[65,72],"yusuf":[83,38],
             "tuğçe":[58,64],"hakan":[65,50],"kağan":[78,36]}

# yukarıda verilen ogrenciler isimli dictionary içerisinde sırayla her öğrencinin
# vize ve final notu yer almaktadır.

# boş bir liste oluşturun
# bu liste içerisine vize'nin %40'ı ile final'in %60'ının toplamı 50'nin altında olan
# kişiler için {isim} KALDI, 50 ve üzeri için {isim} GEÇTİ şeklinde değer ekleyip
# bunları yazdırın

bos_liste = []

for my_key, my_value in ogrenciler.items():
    vize = my_value[0]
    final = my_value[1]

    ortalama = (vize*0.4)+(final*0.6)

    if ortalama>=50:
        bos_liste.append(f"{my_key} {ortalama} notu ile GEÇTİ")
    else:
        bos_liste.append(f"{my_key} {ortalama} notu ile KALDI")

print(bos_liste)
```

```python
sayilar = {"sayi_1":15,"sayi_2":23,"sayi_3":18,"sayi_4":56,
          "sayi_5":20,"sayi_6":26}

# yukarıdaki sayilar isimli dictionary'de yer alan tek sayıları negatifi ile değiştirin
# çift sayıları da yarısını alarak yeni değerini bu şekilde atayın
# bu işlemlerden sonra sayilar isimli dictionary'nin tüm values'leri  değişmiş olacak

# öncelikle sayilar isimli dictionary'nin sadece değerlerini (values) alalım

print(sayilar.values())
print("-"*50)

bos_liste = []

for my_key, my_number in sayilar.items():

    # sayılara eriştikten sonra if ile onun tek mi çift mi olduğunu kontrol ediyoruz
    if my_number%2!=0:
        my_number = -my_number
    else:
        my_number = my_number / 2

    bos_liste.append(my_number)

my_key_list = list(sayilar.keys())

for i in range(len(my_key_list)):
    my_key = my_key_list[i]
    my_value = bos_liste[i]
    sayilar[my_key] = my_value

print(sayilar.values())
```
