```python
# 6. Hafta : GENEL TEKRAR (BUILT-IN DATA STRUCTURES)
# List (Liste) Veri Yapısı
# Köşeli parantez ile tanımlanır. Bu simge, list veri yapısına özel bir simgedir

my_list_1 = []
print(type(my_list_1))
```

```python
# En sık kullanılan List metotları

# Listeye sonradan eleman ekleme
# 1) append() metodu ile
# append() metodu, listenin sonuna elemanı ekler
# yani listenin -1. indeksine eleman ekler
my_list_1 = []

my_list_1.append("Burdur")
my_list_1.append("Mehmet")
my_list_1.append("Akif")
my_list_1.append("Ersoy")
my_list_1.append("Üniversitesi")
# bir listenin tüm elemanlarını [] simgesi içerisinde yan yana yazdırmak için
# liste değişkenin adı yazdırılır. Bu örnekte my_list_1

print(my_list_1)
```

```python
# Eleman ekleme
# 2) insert() metodu ile
# insert() metodunun append metodundan farkı: append() tüm elemanları -1.
# indekse yani en sona eklerdi.
# insert() ise istediğimiz indeks değerine ekler.

# insert() ile -1. indekse eleman ekleme (son eleman)
my_list_1.insert(5,"Bucak")
print(my_list_1)
```

```python
print("-"*50)
# 2. indekse ZTYO ifadesi ekleme

my_list_1.insert(2,"ZTYO")
print(my_list_1)
```

```python
# Listenin elemanlarını kopyalama: copy() metodu ile
# Değer Bazlı kopyalama
list_copy = my_list_1.copy()
```

```python
# Referans bazlı kopyalama
my_list_1_copy = my_list_1
```

```python
print(f"Orijinal Liste = {my_list_1}")
print("-"*50)

my_list_1.append("Kopya Değer")

print(f"Orijinal Liste = {my_list_1}")
print("-"*50)
print(f"copy metodu için liste = {list_copy}")
print("-"*50)
print(f"referans kopyalama işlemi için liste = {my_list_1_copy}")
print("-"*50)
```

```python
# Listeden eleman silme
# 1) remove() metodu ile
# remove() metodu, içerisine değer alır. Bu değer, listenin içinde varsa
# siler

print(f"Orijinal Liste = {list_copy}")
print("-"*50)

my_list_1.remove("Akif")

print(f"remove ile silme sonrası = {my_list_1}")
```

```python
# pop metodu ile istediğiniz indeksteki elemanı silin
# varsayılan silme indeksi = -1

my_list_1.pop(2)

# delete anahtar kelimesinin kısaltması del

# del ile ilk 3 değeri silin (3. değer de dahil)
del my_list_1[:3] # 0,1 ve 2. indeksteki değerleri silecek

# del ile -2. indeksteki elemanı silin
del my_list_1[-2]

print(f"Silme işlemleri sonrası listenin son hali = {my_list_1}")
# del ile direkt listeyi silin
del my_list_1

print(my_list_1)
```

```python
# Listelerde indeksleme işlemi

# liste_adı[tek_değer] => o indeksteki elemanı getirir (tek eleman getirir)
# liste_adı[:bitiş_değeri] => ilk elemandan yani 0. indeksten bitiş_değerine
# kadar olan elemanları getirir (NOT: bitiş_değeri indeksi dahil değildir)
# liste_adı[başlangıç:bitiş] => başlangıç indeksinden bitiş indeksine kadar
# olan elemanları getirir (NOT: bitiş indeksi dahil değildir)

# liste_adı[başlangıç:bitiş:artış] => başlangıç değerinden bitiş değerine
# kadar olan elemanları artış sırasına göre getirir
# NOT liste_adı[::artış] ifadesinde artış miktarı -1 ise
# bu ifade listenin elemanlarını tersten sıralamaya yarar.
# Çünkü -1. indeks son indeks olduğu için onu başlangıç haline getirir ve başa
# doğru bir indeksleme yapar.

my_list_2 = ["Burdur","Mehmet","Akif",18,65,True,123.856,"Ersoy",False,
            "Üniversitesi","Bucak",21,45,True,65.43,1298.65,"ZTYO",False]

eleman_1 = my_list_2[::3]
print(f"my_list_2[::3]'nin 2. elemanı? = {eleman_1[1]}")

print("-"*50)

eleman_1 = my_list_2[2::2]
print(f"my_list_2[2::2]'nin 3. elemanı? = {eleman_1[2]}")

print("-"*50)

my_list_2 = ["Burdur","Mehmet","Akif",18,65,True,123.856,"Ersoy",False,
            "Üniversitesi","Bucak",21,45,True,65.43,1298.65,"ZTYO",False]

# Artış miktarı -1 olduğu için sıralama kesinlikle sondan yani tersten
# BAŞLAYACAK
# Sonrasında, başlangıç değeri olarak 1. indeks verilmiş. Başlagıç indeksi
# her zaman dahildi. Yine dahil ve burada 1. indekste Mehmet elemanı var
# sıralama tersten olduğu için bitiş noktası 0. indeks olacak
# bundan dolayı 1. indeksin solunda sadece 0. indeks olduğu için
# elemanları ['Mehmet','Burdur'] şeklinde indekslendi

eleman_1 = my_list_2[1::-1]
print(f"my_list_2[2::2]'nin 5. elemanı? = {eleman_1}")
```

```python
# for döngüsü kullanarak listenin elemanlarına tek tek erişme

for eleman in my_list_2:
    print(eleman)
```

```python
# boş bir listeye 1'den 100'e kadar olan (100'de dahil) elemanları
# ekleyen kodu yazınız

bos_liste_1 = []

for i in range(1,101):
    bos_liste_1.append(i)

print(bos_liste_1)
```

```python
# iki adet boş liste içerisine, birine 0-100 arasındaki tek sayıları
# diğerine ise çift sayıları ekleyen kodu yazın

liste_tek, liste_cift = [], []
# liste_tek, liste_cift = list(), list()

for sayi in range(101):
    if sayi%2==0:
        liste_cift.append(sayi)
    else:
        liste_tek.append(sayi)

print(f"Çift Sayılar = {liste_cift}")
print("-"*50)
print(f"Tek Sayılar = {liste_tek}")
```

```python
# En sık kullanılan 2. veri yapısı: Dictionary (Sözlük)

# aşağıdaki gibi tanımlanır

# dictionary veri yapısı "key":value şeklinde tanımlanmasıyla diğer tüm
# veri yapılarından apayrı bir yere sahiptir.

my_dict_1 = {"isim":"Furkan","soyisim":"ATLAN","ders_sayısı":5,
            "dersler":["Orta Düzey Programlama","Python Programlama", "OOP"]}

# dictionary.keys() bir dictionary'deki tüm key değerlerini getirir
#
print(f"Sadece keys'ler = {my_dict_1.keys()}")
print("-"*50)
print(f"Sadece values'lar = {my_dict_1.values()}")
print("-"*50)
print(f"hem keys hem de values (birlikte) = {my_dict_1.items()}")

# NOT: eğer items() metodu bir  for döngüsünde kullanılacaksa
# geriye 2 değer döndüreceği için 2 anahtar kelimeye ihtiyaç duyulur

#  örneğin aşağıdaki gibi:

# for deger_1, deger_2 in dictionary.items()
```

```python
# dersler'in sayısını getiren kodu yazın
# herhangi bir değere erişmek istiyorsak aşağıdaki formülü izleriz:
# dictionary_adı['anahtar_adı']

print(f"Değerlerin kendisi = {my_dict_1['dersler']}")
print("-"*50)
print(f"Derslerin Sayısı = {len(my_dict_1['dersler'])}")
```

```python
# bir key'e ait değeri güncellemek
# bunun için aşağıdaki formül izlenir:
# dicitonary_adı['anahtar_adı'] = yeni değer

print(f"Önceki Dersler = {my_dict_1['dersler']}")
print("-"*50)
my_dict_1['dersler'] = ['Orta Düzey Programlama', 'Python Programlama', 'OOP',
                       'Bilgi Güvenliği']

print(f"Şimdiki Dersler = {my_dict_1['dersler']}")
```

```python
# Yeni bir key ve buna ait değerler ekleme

my_dict_1["department"] = "Management Information Systems"

print(my_dict_1.items())
```

```python
# for döngüsü ile hem keys hem de values değerlerine erişim
# her ikisine aynı anda erişmek için .items() fonksiyonu kullanılır
# iki değer döndüğü için bu fonksiyondan, for'dan sonra iki geçici değişken
# tanımlanır

# örnek
for degisken_key, degisken_value in my_dict_1.items():
    print(f"Anahtar = {degisken_key} , Değer = {degisken_value}")

# NOT: Genellikle dictionary veri yapısında şartlar/koşullar işlenirken
# (Özellikle veri bilimi alanında) %99.9 values değerleri üzerinden işlem yapılır
# Ancak, values'ta meydana gelecek değişiklikler izlenebilsin diye onun
# ait olduğu key bilgisi de (adı da) yazdırılır veya işlenir
# Bundan dolayı aynı anda hem key hem de values bilgisine ihtiyac duyulduğu
# için bu tarz işlemlerin %100'ünde dictionary.items() fonksiyonu kullanılır
# for döngüsü ile 2 geçici değişken tanımlanır ve dictionary.items()
# fonksiyonundan gelen 2 değer bu 2 değişkenle tutulur
# if işlemlerinde values değeri kontrol edilir
```

```python
# 1) yukarıdaki dictionary içerisine 'maaslar' isminde bir key ekleyin
# bu key'in değerleri [23000,25000,32000,106000,85000,92000] olacak
# 2) Maaşı 0 ila 25000 arasında olanlara %30, 25001 ila 90000 arasında olanlara
# %20, 90001 ve üzerine %15 zam yapın

# zamlı maaşları, yeni bir listeyle döndürün

my_dict_1["maaslar"] = [23000,25000,32000,106000,85000,92000]
bos_liste = []

my_org_zam_list = my_dict_1["maaslar"]

for my_value in my_org_zam_list:
    if my_value>0 and my_value<=25000:
        bos_liste.append(my_value*1.3)
    elif my_value>25000 and my_value<=90000:
        bos_liste.append(my_value*1.2)
    elif my_value>90001:
        bos_liste.append(my_value*1.15)

print(f"Maaşlar Orijinal = {my_dict_1['maaslar']}")
print("-"*50)
print(f"Maaşlar Zamlı = {bos_liste}")
```

```python
print(f"Maaşlar Orijinal = {my_dict_1['maaslar']}")
print("-"*50)

my_dict_1["maaslar"] = bos_liste

print(f"Maaşlar Zamlı = {my_dict_1['maaslar']}")
```
