```python
# Python'da Fonksiyonlar
# Fonksiyon, verilen komutların tek bir yapı içerisinde çalıştırıldığı
# parametreli-parametresiz, geriye değer döndüren-döndürmeyen olarak
# farklı türleri bulunan yapılardır.
```

```python
#x € Z+ [1,15) f(x) = 6x+4

# Python'da fonksiyonlar def anahtar kelimesiyle tanımlanır
# Söz dizimi aşağıdaki gibidir (parametresiz fonksiyon tanımlamak):

# def fonksiyon_adı():
#     işlemler

# Tanımlanan bir fonksiyon aşağıdaki gibi çalıştırılır (parametresiz fonk için)
# fonksiyon_adı()
```

```python
#x € Z+ [1,15) f(x) = 6x+4
# yukarıda tanımlanan fonksiyonu aşağıdaki gibi kodlayabiliriz
#Parametresiz ve geriye değer döndürmeyen fonksiyon örneği
# Aşağıda, fonksiyonu tanımlıyoruz (henüz çalıştırmadık)
def mat_islemler():
    for sayi in range(1,15):
        print(f"Sayı = {sayi}\t Sonuç = {(sayi*6)+4}")

# Şimdi de fonksiyonumuzu çalıştırıyoruz

mat_islemler()
```

```python
# 1 ile 100 arasında (100'de dahil), tek sayıların karesini,
# çift sayıların küpünü yazdıran fonksiyon
# Not, önce orijinal sayıyı yazdıracak sonra da işlenmiş halini
```

```python
def my_func_1():
    for i in range(1,101):
        if i%2==0:
            print(f"Sayı = {i}\t Sonuç = {i**3}")
        else:
            print(f"Sayı = {i}\t Sonuç = {i**2}")

my_func_1()
```

```python
# 5 ile 500 arasında (500'de dahil) 5'in katı olan tek sayıların
# 7 katının 3 eksiğini, 5'in katı olan çift sayıların 5 katının 3 fazlasını
# hesaplayıp yazdıran fonksiyon

def my_func_2():
    for i in range(5,501,5):
        if i%2==0:
            print(f"Orijinal Sayı = {i}\t Sonuç = {(5*i)+3}")
        else:
            print(f"Orijinal Sayı = {i}\t Sonuç = {(7*i)-3}")

my_func_2()
```

```python
# Parametreli ve geriye değer döndürmeyen fonksiyonlar

# Aşağıdaki söz dizimiyle tanımlanır

# def fonksiyon_adı(parametre_1, parametre_2):
#     işlemler
```

```python
# Dışarıdan verilen listedeki 100'den büyük sayıları yazdıran fonksiyon

# NOT: parametre vereceğiniz zaman isterseniz tipini de belirtebilirsiniz
def my_func_3(liste_deneme:list):
    for eleman in liste_deneme:
        if eleman>100:
            print(eleman)

my_list_1 = [15,80,120,4,-50,130,101,100]

my_func_3(my_list_1)
```

```python
# parametre olarak verilen bir listedeki sadece negatif elemanları yazdıran fonk

def my_func_4(gecici_liste:list):
    for i in gecici_liste:
        if i<0:
            print(i)

my_list_2 = [-15,80,120,4,-50,130,101,-100]

my_func_4(my_list_2)
```

```python
# parametre olarak verilen bir listedeki -50'den büyük negatif tek sayıları
# yazdıran fonk

def my_func_5(gecici_liste:list):
    for sayi in gecici_liste:
        if sayi<0 and sayi%2!=0 and sayi>-50:
            print(sayi)

my_list_3 = [-15,80,120,4,-50,130,-49,-51,-35,-90,-40,101,-100]
my_func_5(my_list_3)
```

```python
# Parametre olarak pozitif bir tamsayı alan ve kendisinden önceki tüm
# pozitif sayıların toplamını yazdıran fonksiyon

# Örneğin; parametre olarak 5 yazdığınızda size 1+2+3+4 işleminin sonucunu 
# yazdıracak

# Çözüm-1

def my_func_5(sayi):
    toplam = 0
    for i in range(sayi):
        toplam = toplam + i
        print(f"Sayı = {i}\t Toplam = {toplam}")
    print(toplam)

my_func_5(5)
print("-"*50)
# Çözüm-2 (Terim sayısı toplamı formülü ile)
def my_func_6(n):
    print(f"Sayı = {n}\t Toplam = {int(((n*(n+1))/2)-n)}")

my_func_6(5)
```

```python
# Faktöriyel bulan fonksiyonu yazın (kendisinden önceki tüm sayıların çarpımı)
# NOT: sayının kendisi de dahil

def my_func_7(sayi):
    faktoriyel = 1
    for i in range(1,sayi+1):
        faktoriyel = faktoriyel * i
        print(f"Sayı = {i}\t Faktöriyel = {faktoriyel}")
    print("-"*50)
    print(f"{sayi} sayısının faktöriyeli = {faktoriyel}")

my_func_7(8)
```

```python
# Parametre olarak bir ifade/metin alan ve bunun tamamını büyük
# harfle yazdıran fonksiyon

def my_func_8(ifade:str):
    print(f"Orijinal = {ifade}\t İşlenmiş = {ifade.upper().replace('I','İ')}")

my_func_8("Furkan")
my_func_8("Burdur Mehmet Akif Ersoy Üniversitesi")
```

```python
# Parametre olarak verilen bir str ifade de kalın ünlüleri
# ince ünlülerle değiştiren fonksiyon (Örneğin, ı değerini i yapacak)

def my_func_9(metin:str):
    print(f"Orijinal = {metin}")
    metin = metin.replace("a","e").replace("ı","i").replace("o","ö").replace("u","ü")
    print(f"İşlenmiş = {metin}")

my_func_9("Burdur Mehmet Akif Ersoy Üniversitesi")
```

```python
# Parametre olarak verilen bir str ifadedeki toplam karakter sayısını
# yazdıran fonksiyon (Örneğin, Furkan ATLAN yazıldığında 12 yazacak)

def my_func_10(ifade:str):
    sayac = 0
    for karakter in ifade:
        sayac = sayac + 1
    print(f"{ifade} ifadesinin karakter sayısı = {sayac}")

my_func_10("Furkan ATLAN")
print(len("Furkan ATLAN"))
```

```python
# Parametre olarak verilen bir str ifadedeki toplam karakter sayısını
# yazdıran fonksiyon (Örneğin, Furkan ATLAN yazıldığında 12 yazacak)
# Boşluk karakteri sayılmayacak

def my_func_11(ifade:str):
    sayac = 0
    for karakter in ifade:
        if karakter != " ":
            sayac = sayac + 1
    print(f"{ifade} ifadesinin karakter sayısı = {sayac}")

my_func_11("Furkan ATLAN")
print(len("Furkan ATLAN"))
```
